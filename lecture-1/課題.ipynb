```python
path = ''
db_name = '課題.db'

try:
    
    # DBオブジェクトの作成
    conn = sqlite3.connect(path + db_name)

    # SQL(RDBを操作するための言語）カーソルオブジェクトを取得
    cur = conn.cursor()

    # SQL文の作成
    # テーブルの作成
    sql = 'CREATE TABLE 課題 (リポジトリ名 TEXT, language TEXT, stars REAL);'

    # SQLの実行
    cur.execute(sql)

except sqlite3.Error as e:
    print("エラーが発生しました", e)

finally:
    # DBへの接軸を閉じる
    conn.close()
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time

base_url = "https://github.com/google"

# もう見たURL
visited = []

# これから見るURL（最初はトップページだけ）
to_visit = [base_url]

# URLとタイトルを保存する辞書
url_title_dict = {}


while to_visit:
    url = to_visit.pop(0)   # 先頭から1つ取り出す

    # 同じURLを何回も見ないように
    if url in visited:
        continue
    visited.append(url)

    # 武蔵野大学のドメインだけ
    if not url.startswith(base_url):
        continue


    # サーバーに優しく
    time.sleep(1)

    try:
        res = requests.get(url, timeout=10)
        res.encoding = res.apparent_encoding
        soup = BeautifulSoup(res.text, "html.parser")

        # タイトル取得
        if soup.title and soup.title.string:
            title = soup.title.string.strip()
        else:
            title = "No Title"

        url_title_dict[url] = title
        print(title, "→", url)

        # ページ内のリンクをさらに追加していく
        for link in soup.find_all("a"):
            href = link.get("href")
            if not href:
                continue

            # 相対パスを絶対URLに
            full_url = urljoin(url, href)

            # ★アンカー(#以降)は削除して同じページ扱いにする
            full_url = full_url.split("#")[0]

            # 武蔵野大学のサイトだけ
            if not full_url.startswith("https://www.musashino-u.ac.jp/"):
                continue

            # もう見た or キューにあるURLはスキップ
            if full_url in visited or full_url in to_visit:
                continue

            to_visit.append(full_url)


    except Exception as e:
        print("⚠️ エラー:", url, e)
```
